```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
marvel = pd.read_csv('superheroes_stats.csv')
marvel.head()
```

```python
marvel.shape
```

```python
marvel['Alignment'].value_counts()
```

```python
plt.figure(figsize=(8,3))
plt.bar(x=marvel['Alignment'].value_counts().index,height=marvel['Alignment'].value_counts())
plt.show()
```

```python
good = marvel[marvel['Alignment'] == 'good']
good.head()
```

```python
good.sort_values(by=['Speed'],ascending=False)
```

```python
max_speed_good = good[good['Speed'] == 100]
max_speed_good
```

```python
min_speed_good = good[good['Speed'] == 1]
min_speed_good
```

```python
good.sort_values(by=['Power'],ascending=False).head()
```

```python
max_power_good = good[good['Power'] == 100]
max_power_good
```

```python
len(max_power_good)
```

```python
max_total_good = good.sort_values(by=['Total'],ascending=False)
max_total_good.head()
```

```python
plt.figure(figsize=(8,5))
plt.bar(x=max_total_good['Name'].head(),height=max_total_good['Total'].head(),color='g')
plt.show()
```

```python
bad = marvel[marvel['Alignment'] == 'bad']
bad.head()
```

```python
bad.sort_values(by=['Intelligence'],ascending=False).head()
```

```python
bad.sort_values(by=['Total'],ascending=False).head()
```

```python
bad_max_total = bad.sort_values(by=['Total'],ascending=False).head()
bad_max_total
```

```python
plt.figure(figsize=(8,5))
plt.bar(x=bad_max_total['Name'].head(),height=bad_max_total['Total'].head(),color=['green','blue','yellow','black','red'])
plt.show()
```
